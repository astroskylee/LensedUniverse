# Environment check

Run this notebook on the server. It will write `requirements_freeze.txt` in the current directory.

```python
import sys
import platform

print("Python:", sys.version)
print("Executable:", sys.executable)
print("Platform:", platform.platform())
```

```python
try:
    import jax
    print("JAX:", jax.__version__)
    print("Devices:", jax.devices())
except Exception as exc:
    print("JAX not available:", exc)
```

```python
import subprocess
from pathlib import Path

cmd = [sys.executable, "-m", "pip", "list", "--format=freeze"]
result = subprocess.run(cmd, capture_output=True, text=True)

if result.returncode != 0:
    print("pip list failed:")
    print(result.stderr)
else:
    output_path = Path("requirements_freeze.txt")
    output_path.write_text(result.stdout)
    print("Wrote", output_path)
    print("Preview:")
    for line in result.stdout.splitlines()[:20]:
        print(line)
```
